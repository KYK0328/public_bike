# 차동차 보유를 1 차량없음을 0으로 변경
자동차본인명을 차량보유로 변경

```python
df["자동차본인명"] = df["자동차본인명"].replace("1;000cc 미만 경차", 1)
df["자동차본인명"] = df["자동차본인명"].replace("1;000cc 이상 ~ 1;600cc 미만", 1)
df["자동차본인명"] = df["자동차본인명"].replace("1;600cc 이상 ~ 2;000cc 미만", 1)
df["자동차본인명"] = df["자동차본인명"].replace("2;000cc 이상", 1)
df["자동차본인명"] = df["자동차본인명"].replace("차량없음", 0)
```

```python
df = df.rename(columns={'자동차본인명':'차량보유'})
```
