# 시각화할때 종류별로 구분하고 시각화

```python
df_문화=df_6[df_6["구분"].str.contains('문화여가시설')]
df_공공=df_6[df_6["구분"].str.contains('공공시설')]
df_교통=df_6[df_6["구분"].str.contains('대중교통')]
```

```python
df_문화
```

```python
#! pip install --upgrade folium 
import folium
from pandas import DataFrame   # 데이터 분석 패키지
from pandas import ExcelFile   # 엑셀파일 가져오기

# 새로운 지도 객체 생성
map_osm = folium.Map(location=[37.273765, 127.002223], zoom_start=12)

for i in df_문화.index:
    name = df_문화.loc[i, '법정동명']
    x = df_문화.loc[i, '위도']
    y = df_문화.loc[i, '경도']
    # 추출한 정보를 지도에 표시
    marker = folium.CircleMarker([x,y],color='yellow', popup=name,radius = 1)
    marker.add_to(map_osm)
for i in df_공공.index:
    name = df_공공.loc[i, '법정동명']
    x = df_공공.loc[i, '위도']
    y = df_공공.loc[i, '경도']
    # 추출한 정보를 지도에 표시
    marker = folium.CircleMarker([x,y],color='blue', popup=name,radius = 1)
    marker.add_to(map_osm)
for i in df_교통.index:
    name = df_교통.loc[i, '법정동명']
    x = df_교통.loc[i, '위도']
    y = df_교통.loc[i, '경도']
    # 추출한 정보를 지도에 표시
    marker = folium.CircleMarker([x,y],color='red', popup=name,radius = 1)
    marker.add_to(map_osm) 
    
map_osm
```
